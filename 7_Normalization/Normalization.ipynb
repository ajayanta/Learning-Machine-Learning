```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\KIIT\Downloads\wine_data.csv",usecols=[0,1,2])
df.columns=['Class Label','Alcohol','Malic acid']
```

```python
df.head()
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['Class Label'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Class Label',axis=1),df['Class Label'],test_size=0.3,random_state=0)
```

```python
x_train.shape,x_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler();
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
#we can check that min  of both column=0 amd max value is 1 
np.round(x_train_scaled.describe(),1)
```

```python
#After scaling the data representation is amost same as before scaling but we can see the scale range [0,1]
fig,axes=plt.subplots(1,2,figsize=(12,5))
sns.scatterplot(x=x_train['Alcohol'],y=x_train["Malic acid"],ax=axes[0])
sns.scatterplot(x=x_train_scaled['Alcohol'],y=x_train_scaled["Malic acid"],ax=axes[1])
axes[0].set_title("Before scaling")
axes[1].set_title("After scaling")
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,5))
sns.kdeplot(x_train['Alcohol'],ax=axes[0])
sns.kdeplot(x_train["Malic acid"],ax=axes[0])
sns.kdeplot(x_train_scaled['Alcohol'],ax=axes[1])
sns.kdeplot(x_train_scaled["Malic acid"],ax=axes[1])
axes[0].set_title("Before scaling")
axes[1].set_title("After scaling")
```
