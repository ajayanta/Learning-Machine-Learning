```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("customer.csv")
```

```python
df.sample(5)
```

```python
df=df[['review','education','purchased']]
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,2],test_size=0.2,random_state=0)
```

```python
y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train);
```

```python
x_trained=oe.transform(x_train)
x_tested=oe.transform(x_test)
```

```python
x_trained=pd.DataFrame(x_trained,columns=x_train.columns)
x_testned=pd.DataFrame(x_tested,columns=x_test.columns)
```
